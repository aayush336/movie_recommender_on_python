````python
 #we need to convert the string of list into list format
import ast
````

````python
# this method will extract the value of name attr of the genres of the movies
# eg {Action,advenure,horror,etc} but here the issue is the values of name are string of list format we need only list
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
````

````python
movies['genres']=movies['genres'].apply(convert)
````

````python
def convert3(obj):
    L =[]
    count = 0
    for i in ast.literal_eval(obj):
        if count!=5:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
````

````python
movies['cast'] = movies['cast'].apply(convert3)
````

````python
def fetch_director(obj):
    L =[]

    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])

    return L
````

now below we are making a stem function and for making that we need a package called as nltk and from it we have imported porterStemmer class

this stem function will help to remove the words with similar meaning <br> eg loved , loving, love will become love,love,love

````python
import nltk
````

````python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
````

````python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
````

````python
new_df['tags'].apply(stem)
````

````python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
````

````python
vectors=cv.fit_transform(new_df['tags']).toarray()
````

````python
cv.get_feature_names_out()
````

now once each movie is represented as a vector <br>
we need to find the distance between them<br>
now as Eucledian distance formula is not a preferable formula to calculate the distance between the vectors<br>
We willl use the cosine distance formula (in which the dist is measured by the angle between them)

````python
from sklearn.metrics.pairwise import cosine_similarity
````

````python
similarity = cosine_similarity(vectors)
````

the above code is good to go with but there are few issues
<br>
when we are calling similarity and den when we sort it the index integretiy will be lost
<br>
eg. similarity func is giving the distance between the selected movie and every other movie in the list. It compares in the order of the movie as they are in the list .

But our main focus is to get the similar movies in recommendation and their postion will not be in any order. So when we call sorted function the integretiy is being lost.<br>
to avoid this **enumarate** func is being called , which maintains the (movie_index,distance) formate and thus it is eassy to trace the movie.

We need it in list form thus list is function is applied
<br>
we will select those whose distance is minimum thus reverse is applied
<br>
- `list(enumerate(similarity[0]))`: Enumerates the elements of `similarity[0]` and converts them into a list. Each element is paired with its corresponding index.
- `sorted(...)`: Sorts the list of tuples obtained from the previous step based on the value of the second element in each tuple.
- `reverse=True`: Specifies that the sorting should be performed in descending order.
- `key = lambda x:x[1]`: Defines the key function used for sorting, which extracts the second element (index 1) from each tuple for comparison.


````python
def recommend(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
````
